```python
from transformers import pipeline
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
classifier = pipeline("sentiment-analysis")
res = classifier("I love you.")
res
```

```python
results = classifier(["i love you.","I hate you", "do you hate me?"])
results
```

```python
model_name = "distilbert-base-uncased-finetuned-sst-2-english" # default model name
```

```python
classifier = pipeline("sentiment-analysis", model = model_name)
res = classifier("I love you.")
res
```

## manual approach

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
save_dir = "saved"
model.save_pretrained(save_dir)
tokenizer.save_pretrained(save_dir)
```

```python
model2 = AutoModelForSequenceClassification.from_pretrained(save_dir)
tokenizer2 = AutoTokenizer.from_pretrained(save_dir)
```

```python
classifier = pipeline("sentiment-analysis", model = model2, tokenizer = tokenizer2)
res = classifier("I love you")
res
```

```python
classifier = pipeline("sentiment-analysis", model = model, tokenizer = tokenizer)
res = classifier("I love you")
res
```

## hindi

```python
# name of pre-trained model on hugging face
model_name = "LondonStory/txlm-roberta-hindi-sentiment"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
classifier = pipeline("sentiment-analysis", model = model, tokenizer = tokenizer)
```

```python
res = classifier(["मुझे लगता है कि मैं तुमसे प्यार करता हूँ लेकिन असल में मैं तुमसे नफरत करता हूँ", "मुझे तुमसे प्यार है", "मुझे आप से नफरत करनी चाहिए"])
print(res) # 0 negative, 1 neutral, 2 positive
```
